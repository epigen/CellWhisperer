```python
from cellwhisperer.utils.model_io import load_cellwhisperer_model
import torch
import pandas as pd

(
pl_model,
tokenizer,
transcriptome_processor,
) = load_cellwhisperer_model(model_path=snakemake.input.model, eval=True, cache=False)

```

```python
query_variants = pd.read_csv(snakemake.input.query_variants, index_col=0)
query_variants
```

```python
text_features = tokenizer(query_variants["variant"].tolist(), return_tensors="pt", padding=True, truncation=True)
text_features
```

```python
_, text_embeds = pl_model.model.get_text_features(**{key: t.to(pl_model.device) for key, t in text_features.items()})
text_embeds
```

```python
torch.save(text_embeds, snakemake.output.text_embeddings)
```
