```python
import pandas as pd
archs4_metadata = pd.read_csv(
    'archs4_sample_metadata.tsv.gz',
    sep = '\t',
    compression = 'gzip'
)
archs4_metadata
```

```python
sra_to_biosample = pd.read_csv(
    'sra_to_biosample.csv.gz',
    compression = 'gzip'
)
sra_to_biosample = sra_to_biosample.dropna().drop_duplicates()
sra_to_biosample.biosample = sra_to_biosample.biosample.astype(int)
sra_to_biosample
```

```python
sample_accessions = pd.read_csv(
    'sample_accessions.tsv.gz',
    sep = '\t',
    compression = 'gzip'
)
sample_accessions = sample_accessions.drop_duplicates()
sample_accessions
```

```python
biosample_metadata = pd.read_csv(
    'biosample_metadata.csv.gz',
    compression = 'gzip'
)
biosample_metadata = biosample_metadata.drop_duplicates()
biosample_metadata
```

```python
metasra_metadata = pd.read_csv(
    'metasra_metadata.csv.gz',
    compression = 'gzip'
)
metasra_metadata = metasra_metadata.drop_duplicates()
metasra_metadata
```

```python
biosample_merged = sra_to_biosample.merge(
    biosample_metadata,
    on = 'biosample',
    how = 'inner'
)
biosample_merged
```

```python
accession_merged = biosample_merged \
    .rename(
        columns = {
            'sra': 'sra_uid',
            'biosample': 'biosample_uid',
            'title': 'biosample_title',
            'attribute': 'raw_biosample_metadata'
        }
    ).merge(
        sample_accessions.rename(
            columns = {
                'uid': 'sra_uid'
            }
        ),
        on = 'sra_uid',
        how = 'inner'
    )
accession_merged
```

```python
metasra_merged = metasra_metadata.merge(
    accession_merged.rename(
        columns = {
            'sample': 'sample_id'
        }
    ),
    on = 'sample_id',
    how = 'inner'
)
metasra_merged
```

```python
archs4_merged = metasra_merged.merge(
    archs4_metadata.drop(
        columns = [
            'srx_accession', 
            'biosample_accession'
        ]
    ).rename(
        columns = {
            'geo_accession': 'geo',
            'title': 'geo_title',
            'source_name_ch1': 'geo_source_name',
            'characteristics_ch1': 'geo_metadata'
        }
    ),
    on = 'geo',
    how = 'inner'
)
archs4_merged
```

```python
def remove_whitespace(string):
    return string.replace(' ', '')


def all_equal(x):
    ref = remove_whitespace(x.iloc[0])
    return all(remove_whitespace(item) == ref for _, item in x.items())


all_equal_idx = archs4_merged[['geo_title', 'biosample_title']].apply(
    all_equal,
    axis = 1
)
```

```python
archs4_merged.loc[
    ~all_equal_idx, 
    ['geo_source_name', 'geo_title', 'biosample_title', 'sample_name', 'biosample', 'experiment']
]
```

```python
from metadatamapping import archs4

archs4_data = archs4.samples(
    'human_gene_v2.2.h5',
    archs4_merged.geo,
    n_processes = 4
)
archs4_data
```

```python
df = archs4_data.obs.merge(
    archs4_merged.set_index('geo'),
    left_index = True,
    right_index = True,
    how = 'left'
).drop_duplicates()
```
