```python
import pandas as pd
archs4_metadata = pd.read_csv(
    '../metadata/archs4_sample_metadata.tsv.gz',
    sep = '\t',
    compression = 'gzip'
)
archs4_metadata
```

```python
sra_to_biosample = pd.read_csv(
    '../metadata/sra_to_biosample.csv.gz',
    compression = 'gzip'
)
sra_to_biosample = sra_to_biosample.dropna().drop_duplicates()
sra_to_biosample.biosample = sra_to_biosample.biosample.astype(int)
sra_to_biosample
```

```python
sample_accessions = pd.read_csv(
    '../metadata/sample_accessions.tsv.gz',
    sep = '\t',
    compression = 'gzip'
)
sample_accessions = sample_accessions.drop_duplicates()
sample_accessions
```

```python
biosample_metadata = pd.read_csv(
    '../metadata/biosample_metadata.csv.gz',
    compression = 'gzip'
)
biosample_metadata = biosample_metadata.drop_duplicates()
biosample_metadata
```

```python
metasra_metadata = pd.read_csv(
    '../metadata/metasra_mapped_metadata.csv.gz',
    compression = 'gzip'
)
metasra_metadata = metasra_metadata.drop_duplicates()
metasra_metadata
```

```python
geo_metadata = pd.read_csv(
    '../metadata/geo_metadata.tsv.gz',
    compression = 'gzip',
    sep = '\t'
)
geo_metadata
```

```python
from metadatamapping import metadata
archs4_annotated = metadata.merge_to_annotated_metadata_frame(
    biosample_metadata,
    sra_to_biosample,
    sample_accessions,
    archs4_metadata,
    geo_metadata
)
archs4_annotated
```

```python
archs4_geo = metadata.annotated_with_metasra_and_treatment(
    archs4_annotated,
    metasra_metadata
)
archs4_geo
```

```python
from metadatamapping import archs4

archs4_data = archs4.samples(
    '../data/human_gene_v2.2.h5',
    archs4_geo.set_index('geo'),
    n_processes = 4
)
archs4_data
```
