```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import os
import sys
import re
```

```python
matplotlib.style.use("/home/moritz/Projects/cellwhisperer/src/plot_style/main.style")
```

```python
df = pd.read_csv(fn)
```

```python
df["model.learning_rate"] = pd.Categorical(df["model.learning_rate"])
df["model.lr_warmup"] = pd.Categorical(df["model.lr_warmup"])
```

```python
fig, ax = plt.subplots(figsize=(2, 2.5))
ax = sns.barplot(data=df, x="model.learning_rate", y="val/loss_epoch", hue="model.lr_warmup", ax=ax)

# move legend to bottom right, without removing the label
ax.legend(loc='upper left', bbox_to_anchor=(1, 1), title="Warmup")
fig.savefig("./hyperparameter_sweep.pdf", bbox_inches="tight")
fig.savefig("./hyperparameter_sweep.png", bbox_inches="tight")
```
